```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import v_measure_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from KMeans import KMeans as my_kmeans
```

# Part 1: EDA

```python
# Load blobs
features, labels = make_blobs(n_samples=1000, n_features=3)
```

## Visualize Expected Cluster 

```python
# Graph the blobs to find initial expected cluster
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')

# Sub dataframe from air_pressure, air_temp, and relative_humidity column
ax.scatter(features[:, 0], features[:, 1], features[:, 2], s=5)
plt.show()

# Here, we can expect our KMeans to capture 3 different cluster.
```

# Part 2 and 3: Implement KMeans in Python and Comparing with Sklearn's KMeans

```python
# Initialize my implementation of kmeans and sklearn's kmeans
cluster_er = my_kmeans(n_clusters=3, max_iter=500,random_state=None)
sklearn_kmeans = KMeans(n_clusters = 3)
```

```python
# Let's run fit on the data
cluster_er.fit(features)
sklearn_kmeans.fit(features)
print(cluster_er.cluster_centers_)
print(sklearn_kmeans.cluster_centers_)

# Looks like they have the same cluster_centers_
```

```python
# Let's graph it
# KMeans implementation
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')

ax.scatter(features[:, 0], features[:, 1], features[:, 2], c = cluster_er.labels_, s=5, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Blobs' KMeans From Scratch")
plt.show()
```

```python
# KMeans sklearn
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')

ax.scatter(features[:, 0], features[:, 1], features[:, 2], c = sklearn_kmeans.labels_, s=5, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Blobs' KMeans From SKLearn")
plt.show()

# The two graph looks identical
```

## Checking v_measure_score of my KMeans and SKlearn's KMeans

```python
# First, let's check on the blobs' dataset
cluster_er.fit(features)
sklearn_kmeans.fit(features)
print(v_measure_score(sklearn_kmeans.labels_, cluster_er.labels_))

# Looks like we're good since this is a perfect score.
# Therefore, I can conclude that in this dataset my kmeans is able to find cluster in the same region. The cluster is also balanced. 
```
